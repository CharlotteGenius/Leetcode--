```python
void backtracking(参数) {
    if (终止条件) {
        存放结果;
        return;
    }

    for (选择：本层集合中元素（树中节点孩子的数量就是集合的大小）) {
        处理节点;
        backtracking(路径，选择列表); // 递归
        回溯，撤销处理结果
    }
}
```

# 77. Combinations
* https://leetcode.com/problems/combinations/description/
* https://programmercarl.com/0077.%E7%BB%84%E5%90%88.html

时间复杂度: O(n * 2^n)

空间复杂度: O(n)

```python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        self.path = []
        self.res = []
        self.backtrack(n, k, 1)
        return self.res

    def backtrack(self, n, k, startind):
        if len(self.path) == k:
            self.res.append(self.path)
            return
        
        for i in range(startind, n+1):
            self.path.append(i) # 处理节点
            self.backtrack(n, k, i+1) # 递归
            self.path = self.path[:-1] # 回溯，撤销处理的节点
```

已经选择的元素个数：path.size();

还需要的元素个数为: k - path.size();

在集合n中至多要从该起始位置 : n - (k - path.size()) + 1，开始遍历，比如1,2,3,4,5里面取3个，那么第二层最后一个为3,4,5，index从3开始就可以，不需要到5

为什么有个+1呢，因为包括起始位置，我们要是一个左闭的集合。

```python
# 优化
# 第二层不需要跑到n+1，而是到 n - (k - len(path)) + 1
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        self.path = []
        self.res = []
        self.backtrack(n, k, 1)
        return self.res

    def backtrack(self, n, k, startind):
        if len(self.path) == k:
            self.res.append(self.path)
            return
        
        for i in range(startind, n-(k-len(self.path))+2):
            self.path.append(i) # 处理节点
            self.backtrack(n, k, i+1) # 递归
            self.path = self.path[:-1] # 回溯，撤销处理的节点
```
